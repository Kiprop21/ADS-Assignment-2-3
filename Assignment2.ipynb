```python
# Import Standard packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# read HR data from csv 
emp = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
emp
```

```python
# Look for missing values
#emp.isnull().sum()
```

```python
# Look for datatypes
#emp.dtypes
```

```python
#Assignment 2 part a
## Show a breakdown of distance from home by job role and attrition.
emp1=emp.groupby(["JobRole",'Attrition'])["DistanceFromHome"].mean().reset_index()
emp1
```

```python
#Assignment 2 part b
#Compare average monthly income by education and attrition.
emp2=emp.groupby(["Education",'Attrition'])["MonthlyIncome"].mean().reset_index()
emp2
```

```python
#Assignment 3 part a
sns.set(style="darkgrid")
```

```python
# Visualize a breakdown of distance from home by job role and attrition.
sns.relplot(x="JobRole", y="DistanceFromHome", data=emp1,col="Attrition")
```

```python
sns.lineplot(data=emp1, x="JobRole", y="DistanceFromHome",hue='Attrition')
```

```python
sns.barplot(x="JobRole", y="DistanceFromHome", hue="Attrition", data=emp1)
```

```python
#Assignment 3 part b
#Visualize average monthly income by education and attrition.
sns.relplot(x="Education", y="MonthlyIncome", data=emp2,col="Attrition")
```

```python
sns.lineplot(data=emp2, x="Education", y="MonthlyIncome",hue='Attrition')
```

```python
sns.barplot(x="Education", y="MonthlyIncome", hue="Attrition", data=emp2)
```
